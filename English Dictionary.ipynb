# English Dictionary

```python
import pickle
def save_Dict(t_d):
    try:
        with open('Dictionary.txt', 'wb') as d:
            pickle.dump(t_d, d)
        print("SUCCESS: writing to dictionary words.txt")
    except Exception as e:
        print("{e} : ERROR")
        print('')


def get_Dict():
    '''
    This function returns a python dictionary,
    after de-serializing the file words.txt
    '''
    try:
        with open('Dictionary.txt', 'rb') as d:
            t_d = pickle.load(d)
        print("SUCCESS: reading from dictionary words.txt")
        return t_d
    except Exception as e:
        print("ERROR: {e}")
        print('')

        
def addNewWord():
    print("__________")
    print("NEW WORD ADDER:")
    print("__________")
    print("Enter a Word to add")
    word = input(">>")
    print(f"Enter a meaning of {word} to add")
    meaning = input(">>")
    d = get_Dict()
    d[word] = meaning
    save_Dict(d)
    print('')
    print('')
    main_menu()
    
    
def search_word():
    print("------------------")
    print("To Find Word:")
    print("------------------")
    word = input('Enter the Word: ')
    d = get_Dict()
    if word in d:
        print('')
        print(f"Found Word:\n[{word}:{d[word]}]")
    else:
        print('Word not found')
        print('')
        search_word()
        
    print('')
    print('')
    main_menu()

def update_record():
    print("------------------")
    print("To Update the Word:")
    print("------------------")
    str1 = input('Please enter the word: ')
    print('')
    d = get_Dict()
    if str1 in d:
        print(f"Found Word:[{str1}:{d[str1]}]")
        print('')
        new_meaning = input('Enter the new meaning: ')
        print('')
        d[str1] = new_meaning
        save_Dict(d)
        
        print(f"SUCCESS: Updated to [{str1} : {d[str1]}]")
        print('')
    else:
        print('Word not found')
        print('')
        update_word()
    print('')
    main_menu()
def main_menu():
    print('')
    print('-----------------------------------------')
    print('--------------------Menu-----------------')
    print('1.Insert a new word')
    print('2.Update a Word')
    print('3.Search for meaning')
    print('4.Exit')
    print('-----------------------------------------')
    print('')
    a = int(input('Please select the number between 1 to 4: '))
    print('')
    if a == 1:
        addNewWord()
    elif a == 2:
        update_record()
    elif a==3:
        search_word()
    elif a==4:
        exit()
        print('Program Exit...:)')
    else:
        print('Error!! Please the number from 1 to 4...!!!')
        main_menu()
    
      
```
